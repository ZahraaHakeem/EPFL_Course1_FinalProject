# C. Exploration: visualizations


## 1

```python
import matplotlib.pyplot as plt
df.reset_index().set_index("Commune")
df= df.sort_values("Residents",ascending=False)
values= df[:10]["Residents"]
colormap = plt.cm.tab10(range(0, len(values)))
plt.xlabel("Population density per km²")
plt.title("Top 10 populated communes in Switzerland")
values.plot.barh(color=colormap);
```

## 2

```python
values= df[:10][["Settlement area","Agricultural area","Wooded area","Unproductive area"]]
values.plot.barh(color=colormap,stacked=True);
plt.legend(bbox_to_anchor=(1.0, 1.0))
plt.title("top 10 populated communes by type of area")
plt.xlabel("area type in %");
```

## 4

```python
import seaborn as sns
sns.violinplot(x="Language", y="0-19 years", data=df, hue="Language")
plt.title("Distribution of age group 0-19 years by language")
plt.xlabel("Language")
plt.ylabel("Percentage of 0-19 years residents")
plt.legend(bbox_to_anchor=(1.0, 1.0));
```

## 5

```python
fig, axi = plt.subplots(nrows = 1,  # sets the number of rows in the subplot grid
                         ncols = 3,  # sets the number of columns in the subplot grid
                         figsize = (12, 3.5) , # sets the width and height of the figure (in inches)
                         sharey=True , # make all subplots share the same y-axix
                         gridspec_kw = {"wspace": 0.2  # Sets the width of padding between subplots (is a fraction of the Axes width)
                                         # Sets relative width ratios between subplots 
                                       }
                        )
ages = ['0-19 years','20-64 years','65 years or over']
for i, ax in zip(range(0, 3), axi.flatten()):    
    sns.violinplot(data=df, x="Language", y=ages[i], ax=axi[i],hue="Language")
    sns.stripplot(x="Language", y=ages[i], data=df,  ax=axi[i], alpha=0.8)
sns.set_style('darkgrid') # 'white', 'dark', 'whitegrid', 'darkgrid', and 'ticks'
plt.show()
```

## 6

```python
types=["Unproductive area", "Wooded area","Settlement area"]
sns.pairplot(df, x_vars=types, 
                   y_vars=["Agricultural area"], 
                   hue="Elevation", 
                   height=3)
plt.title("Relations between Agricultural area and others")
sns.set_style('whitegrid') # 'white', 'dark', 'whitegrid', 'darkgrid', and 'ticks'
plt.show()

# Not linear
```
